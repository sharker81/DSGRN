# DSGRN Python Interface Tutorial
This notebook shows the basics of manipulating DSGRN with the python interface.

````python
import DSGRN
````

## Network
The starting point of the DSGRN analysis is a network specification file. We have a network specification file "network.txt" we will load.

````python
network = DSGRN.Network("network.txt")
````

````python
print(network)
````

````python
print(network.graphviz())
````

## ParameterGraph
Given a network, there is an associated "Parameter Graph", which is a combinatorial representation of parameter space.

````python
parametergraph = DSGRN.ParameterGraph(network)
````

````python
print("There are " + str(parametergraph.size()) + " nodes in the parameter graph.")
````

## Parameter
The `ParameterGraph` class may be regarded as a factory which produces parameter nodes. In the DSGRN code, parameter nodes are referred to simply as "parameters" and are represented as "Parameter" objects.

````python
parameterindex = 34892  # An integer in [0,32592)
````

````python
parameter = parametergraph.parameter(parameterindex)
````

````python
parameter
````

````python
print(parameter)
````

## DomainGraph
Let's compute the dynamics corresponding to this parameter node. In particular, we can instruct DSGRN to create a "domaingraph" object.

````python
domaingraph = DSGRN.DomainGraph(parameter)
````

````python
domaingraph
````

````python
graphviz.Source(domaingraph.graphviz())
````

````python
print(domaingraph.coordinates(5)) # ... I wonder what region in phase space domain 5 corresponds to.
````

## MorseDecomposition
Let's compute the partially ordered set of recurrent components (strongly connected components with an edge) of the domain graph.

````python
morsedecomposition = DSGRN.MorseDecomposition(domaingraph.digraph())
````

````python
graphviz.Source(morsedecomposition.graphviz())
````

How do we get the vertices associated with the Morse set?

````python
[ morsedecomposition.morseset(i) for i in range(0,morsedecomposition.poset().size()) ]
````

## MorseGraph
The final step in our analysis is the production of an annotated Morse graph.

````python
morsegraph = DSGRN.MorseGraph()
````

````python
morsegraph.assign(domaingraph, morsedecomposition)
````

````python
morsegraph
````

````python
print(morsegraph)
````

````python
graphviz.Source(morsegraph.graphviz())
````

# Pattern Matching with Python Interface

The `DomainGraph` class exposes three member functions which are in particular useful for pattern matching.

* `label`
* `direction`
* `regulator`

Current they are documented in `DomainGraph.h`:

```cpp
  /// label
  ///   Given a domain, return a 64-bit integer 
  ///   which indicates whether each wall is an entrance
  ///   wall or an absorbing wall. The information is
  ///   encoded bitwise for the 2*d walls in the following pattern:
  ///     (from least significant to most significant bits)
  ///     left-0, left-1, left-2, ..., left-(d-1),
  ///     right-0, right-1, ... right(d-1)
  ///   A bit of 0 means entrance and 1 means absorbing.
  uint64_t 
  label ( uint64_t domain ) const;

  /// direction
  ///   Given adjacent domains, return the variable
  ///   in which they differ. Given the same domain 
  ///   as both source and target, return dimension().
  ///   Undefined behavior for non-adjacent domains.
  uint64_t
  direction ( uint64_t source, uint64_t target ) const;

  /// regulator
  ///   Given adjacent domains, return the variable
  ///   being regulated between them. Given the same domain 
  ///   as both source and target, return dimension().
  ///   Undefined behavior for non-adjacent domains.
  uint64_t
  regulator ( uint64_t source, uint64_t target ) const;
```

The `label` method can be converted into the "ID?"-type string which tells us for each dimension if the domain supports trajectories which are guaranteed increasing ("I"), guaranteed decreasing ("D"), or not guaranteed to be increasing or decreasing ("?"). 

Here is python code to perform this conversion:

````python
def labelstring(L,D):
  """
  Inputs: label L, dimension D
  Outputs:"label" output L of DomainGraph is converted into a string with "I", "D", and "?"
  """
  return ''.join([ "D" if L&(1<<d) else ("I" if L&(1<<(d+D)) else "?") for d in range(0,D) ])
````

Here is an example of its use:

````python
{ v : labelstring(domaingraph.label(v),domaingraph.dimension()) for v in range(0,domaingraph.digraph().size())}
````

Assume we want to restrict this to Morse set 1:

````python
morseset_of_interest = morsedecomposition.morseset(1)
````

````python
{ v : labelstring(domaingraph.label(v),domaingraph.dimension()) for v in morseset_of_interest}
````
